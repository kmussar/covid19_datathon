```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Race_data_tracker.csv')
df.info()
df.head()
```

```python
df = df.groupby(by='state').sum().reset_index()
df.info()
df.head()
```

```python
df.columns
```

```python
df.sort_values(by='Positives_ race_Black ')
black = df[['state', 'Positives_Total','Positives_ race_Black ', 'Deaths_race_Black ']]
black.head()
```

```python
black.dropna()
black = black[black['Positives_ race_Black '] > 0]
black
```

```python
black['mortality_rate'] = (black['Deaths_race_Black ']/black['Positives_ race_Black ']*100).round(2)
black['%positives_black'] = (black['Positives_ race_Black ']/black['Positives_Total']*100).round(2)
black.head()
```

```python
columns = black.columns
```

```python
black.columns
```

```python
black.sort_values(by='%positives_black',ascending=False,inplace=True)
black
```

## Merge onto CCVI data - statewide

```python
ccvi = pd.read_csv('state_ccvi.csv')
ccvi.info()
ccvi.head()
```

```python
black_ccvi = black.join(ccvi.set_index('State Abbreviation'), on='state', how='left', sort=True, lsuffix=1)
black_ccvi.info()
black_ccvi.head()
```

```python
black.columns
```

```python
new_columns = (['state_abbrev', 'pos_total', 'pos_black',
       'deaths_black ', 'mortality_rate', '%pos_black',
        'state', 'socioeconomic', 'household_and_disability', 
        'minority', 'housing_and_transport', 'epidemiological', 
        'healthcare', 'ccvi'])
black_ccvi.rename(columns=dict(zip(black_ccvi.columns,new_columns)),inplace=True)
black_ccvi.head()
```

# Linear Regression Model
Train model only - no testing 
Using model to pick out the strongest correlations 


```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_val_predict
import scipy.stats as stats
```

```python
black_ccvi.columns
```

```python
X = black_ccvi.drop(columns=['state_abbrev','pos_total', 'pos_black', 'deaths_black ',
       'mortality_rate', '%pos_black', 'state'],axis=1)
y = black_ccvi['mortality_rate']
```

```python
lr = LinearRegression()
lr.fit(X, y)
predictions = lr.predict(X)

print('MAE:', metrics.mean_absolute_error(y, predictions))
print('MSE:', metrics.mean_squared_error(y, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y, predictions)))
print('R^2:', metrics.explained_variance_score(y,predictions))
```

# Next, look at the overall mortality rate of blacks vs whites

### This table includes all races - positives, negatives, and deaths. Combined with CCVI data

```python
race_ccvi = df.join(ccvi.set_index('State Abbreviation'), on='state', how='left', sort=True, lsuffix=1)
race_ccvi.info()
race_ccvi.head()
```

### Calculate mortality rate and % positives for each race

```python
race_ccvi.columns
```

```python
positives = (['Positives_Total', 'Positives_race_White ',
       'Positives_ race_Black ', 'Positives_race_ LatinX _Hispanic',
       'Positives_ race_Asian', 'Positives_ race_AIAN', 'Positives_ race_NHPI',
       'Positives_ race_Multiracial ', 'Positives_race_Other',
       'Positives_race_Unknown'])
deaths = (['Deaths_Total', 'Deaths_race_White ', 'Deaths_race_Black ',
       'Deaths_race_ LatinX_Hispanic', 'Deaths_race_Asian', 'Deaths_race_AIAN',
       'Deaths_race_NHPI', 'Deaths_race_Multiracial ', 'Deaths_race_Other',
       'Deaths_race_Unknown'])
mortality_rate_columns = (['mortality_rate_total', 'mortality_rate_white','mortality_rate_black',
                         'mortality_rate_latinx','mortality_rate_asian', 'mortality_rate_AIAN'
                          'mortality_rate_NHPI', 'mortality_rate_multiracial','mortality_rate_other',
                          'mortality_rate_unknown'])
pos_columns = (['%pos_total', '%pos_white','%pos_black',
                         '%pos_latinx','%pos_asian', '%pos_AIAN'
                          '%pos_NHPI', '%pos_multiracial','%pos_other',
                          '%pos_unknown'])
```

```python
for i in range(len(positives)+1): 
    race_ccvi[mortality_rate_columns[i]] = (race_ccvi[deaths[i]]/race_ccvi[positives[i]]*100).round(2)
    #race_ccvi[pos_columns[i]] = (race_ccvi[positives[i]]/race_ccvi['Positives_Total'][0]*100).round(2)

race_ccvi.head()
```

```python
race_ccvi['mortality_rate_total'] = (race_ccvi['Deaths_Total']/race_ccvi['Positives_Total']*100).round(2)
race_ccvi.head()
```

```python
race_ccvi[['state', 'mortality_rate_total']]
```

```python
# drop any rows where there are no positives 
race_ccvi = race_ccvi[race_ccvi['Positives_Total']>0]
race_ccvi.head()
```

```python
race_ccvi['mortality_rate_white'] = (race_ccvi['Deaths_race_White ']/race_ccvi['Positives_race_White ']*100).round(2)
race_ccvi['%positives_white'] = (race_ccvi['Positives_race_White ']/race_ccvi['Positives_Total']*100).round(2)

race_ccvi['mortality_rate_black'] = (race_ccvi['Deaths_race_Black ']/race_ccvi['Positives_ race_Black ']*100).round(2)
race_ccvi['%positives_black'] = (race_ccvi['Positives_ race_Black ']/race_ccvi['Positives_Total']*100).round(2)

race_ccvi.head()
```
