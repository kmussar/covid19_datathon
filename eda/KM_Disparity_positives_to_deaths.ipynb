```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Race_data_tracker.csv')
df.info()
df.head()
```

```python
df = df.groupby(by='state').sum().reset_index()
df.info()
df.head()
```

```python
df.columns
```

```python
df.sort_values(by='Positives_ race_Black ')
black = df[['state', 'Positives_Total','Positives_ race_Black ', 'Deaths_race_Black ']]
black.head()
```

```python
black.dropna()
black = black[black['Positives_ race_Black '] > 0]
black
```

```python
black['mortality_rate'] = (black['Deaths_race_Black ']/black['Positives_ race_Black ']*100).round(2)
black['%positives_black'] = (black['Positives_ race_Black ']/black['Positives_Total']*100).round(2)
black.head()
```
