# Case Study
### Non-Hispanic Black or African American
* **South Carolina** - 53.5% of COVID-19 deaths; 30% of population
* **Tennessee** - 28.2% of COVID-19 deaths; 33.2% of population

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('county_detailed_demographics_and_deaths.csv')
df.info()
df.head()
```

```python
# create dataframe for TN 
tn = df[df['State Abbreviation'] == 'TN']
tn.head()
```

```python
# create dataframe for SC 
sc = df[df['State Abbreviation'] == 'SC']
sc.head()
```

```python
sc_corr_matrix = sc.corr()
sc_corr_matrix['deaths'] = sc_corr_matrix['deaths'].abs()
sc_corr_matrix.sort_values(by='deaths', ascending=False, inplace=True)
sc_corr_matrix.reset_index(inplace=True)
sc_corr_matrix[['index','deaths']]
```

```python
tn_corr_matrix = tn.corr()
tn_corr_matrix['deaths'] = tn_corr_matrix['deaths'].abs()
tn_corr_matrix.sort_values(by='deaths', ascending=False, inplace=True)
tn_corr_matrix.reset_index(inplace=True)
tn_corr_matrix[['index','deaths']]
```
